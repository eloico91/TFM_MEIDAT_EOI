## Eliminamos Variables

```python
borrar = df.loc[:,df.columns.str.contains('Date|Measure|Temperature|Humidity')].columns
df.drop(columns = borrar,inplace = True)
```

```python
df.info()
```

```python
df
```

## Creamos variables sintéticas

En primer lugar claculamos el tiempo total de producción

```python
df['Tiempo_Total'] = df.index - df.index[0]
```

```python
from datetime import timedelta
df['Tiempo_Total_min'] = df['Tiempo_Total'].apply(lambda x: x.total_seconds()/60)
```

Creamos un aleatorio de tarros cerrados y tarros defectuosos

```python
import random
df['Tarros_Cerrados_aleatorios'] = [random.randrange(6000,9000) for p in range(0, len(df))]
df['Tarros_Defectuosos'] = [random.randrange(60,500) for p in range(0, len(df))]
df['Tarros_Defectuosos'][0] = 0
```

Creamos la variable de Tiempo en funcionamiento de la máquina

```python
df['Tiempo_Total_min_lag_1'] = df.Tiempo_Total_min.shift(1, fill_value=0)
df['Tiempo_funcionamiento_min'] = df['Tiempo_Total_min'] - df['Tiempo_Total_min_lag_1']
df.drop(columns = ['Tiempo_Total_min_lag_1','Tiempo_Total_min','Tiempo_Total'],inplace = True)
```

Introducimos la condicion simple de que cuando da fallos no se cierran tarros

```python
df['Tarros_Cerrados'] = np.where(df['Failure'] == 'Yes',0,df['Tarros_Cerrados_aleatorios'])
df.drop(columns = 'Tarros_Cerrados_aleatorios',inplace = True)
```

Introducimos la condición simple de que cuando hay fallo hay un tiempo de parada

```python
df['Tiempo_Paradas_min'] = np.where(df['Failure'] == 'Yes', df['Tiempo_funcionamiento_min'],0)
df['Tiempo_Producción_min'] = df['Tiempo_funcionamiento_min'] - df['Tiempo_Paradas_min']
```

```python
df.head()
```

Introducimos una condición múltiple que para mayor tiempo de producción mayor es el número de tarros cerrados

```python
#Lista de condiciones
condiciones = [(df['Tiempo_Producción_min'] < 60),
               (df['Tiempo_Producción_min'] == 60),
               (df['Tiempo_Producción_min'] > 60)]

#Lista de resultados
resultados = [0,df['Tarros_Cerrados'],df['Tarros_Cerrados']*df['Tiempo_Producción_min']/60]
                
#Aplicamos select
df['Tarros_Cerrados'] = np.select(condiciones,resultados, default = -999)
```

Añadimos una variable del máximo de producción posible con el redimiento de la máquina

```python
df['Tarros_max_prod'] = df['Tiempo_funcionamiento_min'] * 2.5 *60
```

Pasamos el tiempo desde la última avería a minutos

```python
df['Tiempo_ultima_averia_min'] = df['Hours Since Previous Failure'] * 60
```

Creamos la variabel de tarros correctos

```python
df['Tarros_correctos'] = df['Tarros_Cerrados'] - df['Tarros_Defectuosos']
```

Ordenamos el dataset

```python
df = df.loc[:,['Operator','Failure','Tarros_max_prod','Tarros_Cerrados','Tarros_Defectuosos','Tarros_correctos','Tiempo_funcionamiento_min','Tiempo_Paradas_min','Tiempo_Producción_min','Tiempo_ultima_averia_min']]
df
```

Aplicamos una condición simple de que cuando hay fallo los Tarros correctos no son negativos sino 0

```python
df['Tarros_correctos'] = np.where(df['Tarros_correctos'] < 0,0,df['Tarros_correctos'])
```

```python
df
```

Generamos un tiempo de funcionamiento más realista.

```python
df['Tiempo_Producción_min'] = df['Tiempo_Producción_min']*[random.uniform(0.75, 0.95) for p in range(0, len(df))]
```
