```python
import cloudpickle
import pandas as pd
from janitor import clean_names

ruta_proyecto = 'C:/Users/eloic/Google Drive/EOI/TFM//01_MANT_PREDICTIVO'

nombre_fichero_datos = 'validacion.csv'

ruta_completa = ruta_proyecto + '/02_Datos/02_Validacion/' + nombre_fichero_datos

df = pd.read_csv(ruta_completa,index_col= 'Date', sep = ',')

df.rename(columns = {'Temperature':'temperatura',
                     'Humidity':'humedad',
                     'Hours Since Previous Failure':'horas_desde_utlimo_fallo',
                     'Failure':'fallo'}, inplace=True)

df.drop_duplicates(inplace = True)

variables_finales = ['horas_desde_utlimo_fallo',
                     'humedad',
                     'temperatura']

df = df[variables_finales].copy()

nombre_pipe_ejecucion = 'pipe_ejecucion.pickle'

ruta_pipe_ejecucion = ruta_proyecto + '/04_Modelos/' + nombre_pipe_ejecucion

with open(ruta_pipe_ejecucion, mode='rb') as file:
   pipe_ejecucion = cloudpickle.load(file)

pred = pipe_ejecucion.predict_proba(df)[:, 1]
```
