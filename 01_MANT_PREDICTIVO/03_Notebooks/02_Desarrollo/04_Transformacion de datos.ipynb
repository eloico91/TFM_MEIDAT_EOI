## TRANSFORMACIÓN DE CATEGÓRICAS

### One Hot Encoding

#### Variables a aplicar OHE

```python
var_ohe = ['operador', 'medida_2', 'medida_3', 'fallo', 'mes', 'dia_mes','dia_semana']
```

#### Instanciar

```python
ohe = OneHotEncoder(sparse = False, handle_unknown='ignore')
```

#### Entrenar y aplicar

```python
cat_ohe = ohe.fit_transform(cat[var_ohe])
```

#### Guardar como dataframe

```python
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out())
```

## TRANSFORMACIÓN DE NUMÉRICAS

## UNIFICAR DATASETS TRANSFORMADOS

```python
df = pd.concat([cat_ohe,num.reset_index()],axis = 1).set_index('Date')
```

```python
df.columns
```

## REESCALAR VARIABLES

### Con Min-Max

```python
df.info()
```

Seleccionamos todas variables menos en las que hemos hecho One Hot Encoding.

#### Variables a reescalar con Min-Max

```python
var_mms = df.iloc[:,68:].columns#En este caso desde la 67
```

```python
var_mms
```

#### Instanciar

```python
mms = MinMaxScaler()
```

#### Entrenar y aplicar

```python
df_mms = mms.fit_transform(df[var_mms])
```

#### Guardar como dataframe

```python
#Añadir sufijos a los nombres
nombres_mms = [variable + '_mms' for variable in var_mms]

#Guardar como dataframe
df_mms = pd.DataFrame(df_mms,columns = nombres_mms)
```
